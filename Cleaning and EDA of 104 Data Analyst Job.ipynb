# Data Cleaning
drop duplicates, calculate average salary, split location, change experience_req to numeric, analyzing skills in description

```python
df = df.drop(['Unnamed: 0'], axis = 1).copy()
```

```python
df
```

```python
# Check duplicates
df.loc[df.duplicated()]
```

```python
df = df.drop_duplicates().reset_index(drop=True).copy()
df
```

```python
df['location'].value_counts()
```

```python
# Split location
city = []
district = []
for l in df['location']:
    cit = l[:3]
    dis = l[3:]
    
    city.append(cit)
    district.append(dis)

df['city'] = city
df['district'] = district

df
```

```python
df = df.drop('location', axis=1).copy()
df
```

```python
df.head()
```

```python
# Extract number from experience_req and change to interger
import re
experience_req = []
for e in df['experience_req']:
    match = re.search(r'(\d+)', e)
    if match:
        exp = int(match.group(1))
    else:
        exp = 0
    experience_req.append(exp)
    
df['experience_req'] = experience_req
```

```python
# Calculate average_salary and change to float type
average_salary = []
for s in df['salary']:
    match_mon = re.search(r'月薪([\d,]+)~([\d,]+)元', s)
    match_above = re.search(r'月薪([\d,]+)元以上', s)
    if match_mon:
        avg_s = int((int(match_mon.group(1).replace(',', '')) + int(match_mon.group(2).replace(',', ''))) / 2)
    elif match_above:
        avg_s = int(match_above.group(1).replace(',', ''))
    else:
        avg_s = 'NaN'
    average_salary.append(avg_s)
```

```python
df['average_salary'] = average_salary
df = df.drop('salary', axis=1)
```

```python
df['average_salary'] = df['average_salary'].replace('NaN', np.nan)
df['average_salary'] = df['average_salary'].astype('float')
```

```python
# Extract frequently mentioned skills
from collections import Counter


# Preprocess the text (convert to lowercase, tokenize)
words_list = []
for description in df['description']:
    description = description.lower()
    words = re.findall(r'\b\w+\b', description)
    power_bi_words = re.findall(r'\bpower bi\b', description)
    machine_learning_words = re.findall(r'\bmachine learning\b', description)
    looker_studio_words = re.findall(r'\blooker studio\b', description)
    cleaning_words = re.findall(r'清理', description)
    modeling_words = re.findall(r'建模', description)
    words_list.extend(words)
    words_list.extend(power_bi_words)
    words_list.extend(machine_learning_words)
    words_list.extend(looker_studio_words)
    words_list.extend(cleaning_words)
    words_list.extend(modeling_words)


# Remove common stop words
# stop_words = ["and", "in", "with", "required", '和']

# filtered_text = [word for word in processed_text if word not in stop_words]

# Count term frequencies
words_frequencies = Counter(words_list)


# Define a list of skills to extract
skills_to_extract = ["sql", "python", "tableau", "excel", 'power bi', 'looker studio', 'r', 'machine learning', '清理', '建模']

# Find frequencies of skills
skill_frequencies = {}
for skill in skills_to_extract:
    frequency = words_frequencies.get(skill, 0)
    skill_frequencies[skill] = frequency

# Sort skills by frequency (most frequent first)
sorted_skills = sorted(skill_frequencies.items(), key=lambda x: x[1], reverse=True)

# Print the most frequently mentioned skills
for skill, frequency in sorted_skills:
    print(f"Skill: {skill}, Frequency: {frequency}")
```
